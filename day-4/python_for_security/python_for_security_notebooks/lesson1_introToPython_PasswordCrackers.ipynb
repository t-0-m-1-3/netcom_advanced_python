Morris infected 6,000 computers with a simple worm, that would take advantage of the sendmail program in Unix. It then exploited a vulnerability in the `finger` daemon and finally used a remote shell to connect to the target. If any one of these 3 attack vectors succeeded, the worm would pull the rest of the virus over.
